```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('train.csv')

df.head()
```

```python
df.shape
```

```python
df["fuel_type"].value_counts()
```

```python
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Avenir Next LT Pro'
brand_counts = df['brand'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,7))

bar_color = '#F5A700'

sns.barplot(x=brand_counts.index, y=brand_counts.values, edgecolor=bar_color, color=bar_color)

plt.title('Distribution of Vehicle Brands', fontsize=18, weight='bold', pad=20, color='#333333')

plt.xlabel('Brand', fontsize=14, weight='bold', color='#333333')
plt.ylabel('Count', fontsize=14, weight='bold', color='#333333')

plt.xticks(rotation=90, fontsize=12, ha='center', color='#333333')

plt.yticks(fontsize=12, color='#333333')

plt.gca().set_facecolor('#F7F7F7')

plt.grid(axis='y', linestyle='--', alpha=0.5)

sns.despine()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fuel_counts = df['fuel_type'].value_counts()

plt.figure(figsize=(10, 8))

explode = [0.1 if count == fuel_counts.max() else 0 for count in fuel_counts]

plt.pie(
    fuel_counts,
    explode=explode,
    startangle=140,
    colors=sns.color_palette("Set3"),
    autopct=None
    )

plt.title('Fuel Type Distribution', fontsize=15)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='accident', data=df, palette='Set2')
plt.title('Accident Status Proportion', fontsize=15)
plt.xlabel('Accident Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
